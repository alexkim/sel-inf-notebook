```python
import numpy as np
import pandas as pd
import feather
import json
```

```python
# Generate a sample array of floats
n = 50
p = 2

sample_data = np.random.normal(size=(n,p))  # generate data
sample_data = pd.DataFrame(sample_data)  # cast to pandas
```

```python
def float_array_to_json(float_array):
    """Save an array of floats to JSON.
    """
    # Write the array to a temporary file
    filepath = 'float_array_temp'
    feather.write_dataframe(sample_data, filepath)
    
    # Read the temporary file as bytes
    array_data = open(filepath, 'rb').read()
    
    # Convert raw bytes to a list of ints
    array_bytes_as_ints = []
    for d in array_data:
        array_bytes_as_ints.append(d)
    
    # Return the JSON representation of the list of ints
    return json.dumps(array_bytes_as_ints)

float_array_to_json(sample_data)
```
